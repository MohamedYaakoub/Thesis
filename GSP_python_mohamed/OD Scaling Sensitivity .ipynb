```python
def create_barplot(array1, array2, title):
    # Sample data: replace with your own values
    labels = ["TT25", "TT3", "Ps3", "TT49", "Wf", "N2"]

    # Calculate the bar positions
    bar_positions = np.arange(len(labels))

    # Create a figure and a set of subplots
    fig, ax = plt.subplots(figsize=(10, 4), dpi=500)

    # Create the bar graph with bars centered on the labels
    bar_width = 0.1
    ax.bar(bar_positions - bar_width/2, array1, bar_width, label=r'$C_V$ = 1.05', color='#009ADE', edgecolor='black')
    ax.bar(bar_positions + bar_width/2, array2, bar_width, label=r'$C_V$'' = 0.95', color='#00CD6C', edgecolor='black')
    
    # Set labels for the x-axis and rotate them for better readability
    ax.set_xticks(bar_positions)
    ax.set_xticklabels(labels)
    plt.xticks(rotation=0)

    # Add labels and a title
    ax.set_xlabel('Categories')
    ax.set_ylabel('Change from design point conditions [%]')
#     ax.set_title('Bar Graph with Centered Bars')

    # Display the legend
    ax.legend()

    # Show the graph
    ax.yaxis.grid(alpha=0.7)  # grid lines
    ax.set_axisbelow(True)  # grid lines are behind the res
    
    lower_lim = np.min([array1, array2])
    upper_lim = np.max([array1, array2])
    
    if abs(lower_lim) > abs(upper_lim):
         upper_lim = abs(lower_lim)
    else:
        lower_lim = -abs(upper_lim)
    
    
    plt.ylim([lower_lim - 0.1, upper_lim + 0.1])
    plt.tight_layout()
    plt.savefig(f"C:/Users/mohsy/University/KLM/Thesis/My thesis/Plots/Report plots/{title}.jpeg")
    plt.show()

create_barplot(((y_sim_CV_c_up - y_sim_og)/y_sim_og)[0] * 100, ((y_sim_CV_c_down - y_sim_og)/y_sim_og)[0] * 100, 'Core nozzle CV sensitivity')
create_barplot(((y_sim_CV_d_up - y_sim_og)/y_sim_og)[0] * 100, ((y_sim_CV_d_down - y_sim_og)/y_sim_og)[0] * 100, 'Bypass nozzle CV sensitivity')

print(y_sim_CV_d_up)
print(y_sim_og)
# print(y_sim_CX_c_up - y_sim_og)
# print(y_sim_CX_c_down - y_sim_og)
# print(y_sim_CX_d_up - y_sim_og)
# print(y_sim_CX_d_down - y_sim_og)
```

```python
def plot_mapT(file_name, DP_SF_PR, DP_SF_Wc, DP_SF_Eta, DP_SF_Nc, PRs, WCs, Etas):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 8), dpi=500, sharex='col')
    
    PRminR, PRmaxR, MdotTR, EtaTR, NPrTR, NTR, B = read_mapT(file_name, file_name, 0)
    PRmin, PRmax, MdotT, EtaT, NPrT, NT, B = read_mapT(file_name, file_name, 1)
    # extraxt the iso speed lines
    no_betalines = len(B)
    listsER = np.array([EtaTR[x:x + no_betalines] for x in range(0, len(EtaTR), no_betalines)], dtype=object)
    listsMR = np.array([MdotTR[x:x + no_betalines] for x in range(0, len(MdotTR), no_betalines)], dtype=object)
    listsE  = np.array([EtaT[x:x + no_betalines] for x in range(0, len(EtaTR), no_betalines)], dtype=object)
    listsM  = np.array([MdotT[x:x + no_betalines] for x in range(0, len(MdotTR), no_betalines)], dtype=object)

    listsPR_R = np.zeros(listsER.shape)
    listsPR   = np.zeros(listsE.shape)

    for i in range(len(PRmax) - 1):
        p_ref  = np.polyfit([0, 1], [PRminR[i + 1], PRmaxR[i + 1]], 1)
        p      = np.polyfit([0, 1], [PRmin[i + 1], PRmax[i + 1]], 1)
        PR_ref = np.polyval(p_ref, B)
        PR     = np.polyval(p, B)
        listsPR_R[i, 0:no_betalines] = PR_ref
        listsPR[i, 0:no_betalines]   = PR
    
    listsPR = DP_SF_PR * (listsPR - 1) + 1
    listsM = DP_SF_Wc * listsM
    listsE  = DP_SF_Eta * listsE
    
    # plotting
    
    iter_length = int(len(listsER)/len(np.unique(listsER)))
#     for SF_PR, SF_Wc, SF_Eta, color, legend in zip(OD_SF_PR, OD_SF_Wc, OD_SF_Eta, ['#FF1F5B', '#009ADE', '#00CD6C'],
#                                                   ['Take off', 'Climb', 'Cruise']):

    plot_PR =  listsPR 
    plot_Wc =  listsM
    plot_Eta = listsE
    color = '#80102e'
    for i in range(len(listsER)):
#             ax1.plot(listsPR_R[i], listsER[i], linestyle="-.", c='grey')  # plot the reference map
        ax1.plot(plot_PR[i], plot_Wc[i], c=color, linewidth=1)  # plot the modified map

#             ax2.plot(listsPR_R[i], listsMR[i], linestyle="-.", c='grey')  # plot the reference map
        ax2.plot(plot_PR[i], plot_Eta[i], c=color, linewidth=1)  # plot the modified map

    for i, (color, legend) in enumerate(zip(['#009ADE', '#FF1F5B', '#00CD6C'],
                                     [r'$CV$ = 1.05', r'$CV$ = 1', r'$CV$ = 0.95'])):
        ax1.scatter(y=WCs[i], x=PRs[i], c=color, edgecolor='black', alpha=0.7, label=legend)
        ax2.scatter(y=Etas[i], x=PRs[i], c=color, edgecolor='black', alpha=0.7, label=legend)

    ax1.set_xlabel("Pressure Ratio")
    ax1.set_ylabel("Mass flow")
    ax1.grid()
    ax1.set_title(file_name)
    ax2.set_xlabel("Pressure Ratio")
    ax2.set_ylabel("Efficiency")
    ax2.grid()
    ax2.set_title(file_name)
    plt.tight_layout()
    plt.show()
```
