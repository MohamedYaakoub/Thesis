```python
import numpy as np
import seaborn as sns
import pickle
from matplotlib import pyplot as plt
from matplotlib.ticker import ScalarFormatter
```

```python
file_name = "Reynolds_CEOD_data_mohamed_2019_feb_1-9_2"
GEnx_OD, GEnx_OD_true, _, alt_time, All_Reynolds = pickle.load(open(f"Reynolds_pickle/{file_name}.p", "rb"))
take_off_All_Reynolds, climb_All_Reynolds, cruise_All_Reynolds = All_Reynolds
Re2, Re25, Re3, Re4, Re49, Re5, Re14, Re19, _, _ = np.array(take_off_All_Reynolds).T
```

```python
fig_title = "Corrected N1 speed"

fig, ax = plt.subplots(1, 2, figsize=(8, 4), dpi=500)
for flight_phase, legend in zip(range(3), ['Take off', 'Climb', 'Cruise']):

    ax[0].scatter(y=GEnx_OD[flight_phase][:, 0], x=alt_time[flight_phase][:, 1], s=2, alpha=0.7,label=legend)
    ax[1].scatter(y=GEnx_OD[flight_phase][:, 0], x=alt_time[flight_phase][:, 0], s=2, alpha=0.7,label=legend)

    ax[0].set_ylabel(f'Corrected N1 speed to station 12 [%]')
    ax[0].set_xlabel('Time [s]')
    
    ax[1].set_ylabel(f'Corrected N1 speed to station 12 [%]')
    ax[1].set_xlabel('Altitude [m]')
    
    
    
    
ax[0].grid(True, alpha=0.2)
ax[1].grid(True, alpha=0.2)
# ax.set_title(title)

# Use ScalarFormatter for the y-axis
x_formatter = ScalarFormatter(useMathText=True)
x_formatter.set_scientific(True)
x_formatter.set_powerlimits((-1, 1))  # Adjust the power limits for x-axis

y_formatter = ScalarFormatter(useMathText=True)
y_formatter.set_scientific(True)
y_formatter.set_powerlimits((-1, 1))  # Adjust the power limits for y-axis

ax[0].xaxis.set_major_formatter(x_formatter)
ax[0].yaxis.set_major_formatter(y_formatter)
ax[1].xaxis.set_major_formatter(x_formatter)
ax[1].yaxis.set_major_formatter(y_formatter)
fig.tight_layout()
fig.legend(*fig.axes[0].get_legend_handles_labels(), loc='upper right')
plt.savefig("C:/Users/mohsy/University/KLM/Thesis/My thesis/Plots/Report plots/" 
           + fig_title + '.jpeg')
```

```python
fig_title = "Flight Overview"

fig, ax = plt.subplots(1, 1, figsize=(8, 6), dpi=500)
for flight_phase, legend in zip(range(3), ['Take off', 'Climb', 'Cruise']):
    ax.scatter(alt_time[flight_phase][:, 1], alt_time[flight_phase][:, 0], s=2, alpha=0.7,label=legend)
#     ax.plot(alt_time[flight_phase][:, 1], alt_time[flight_phase][:, 0], label=legend)


ax.set_xlabel(f'Offset')
ax.set_ylabel('Altitude [m]')
ax.grid(True, alpha=0.2)
# ax.set_title(title)

# Use ScalarFormatter for the y-axis
x_formatter = ScalarFormatter(useMathText=True)
x_formatter.set_scientific(True)
x_formatter.set_powerlimits((-1, 1))  # Adjust the power limits for x-axis

y_formatter = ScalarFormatter(useMathText=True)
y_formatter.set_scientific(True)
y_formatter.set_powerlimits((-1, 1))  # Adjust the power limits for y-axis

ax.xaxis.set_major_formatter(x_formatter)
ax.yaxis.set_major_formatter(y_formatter)
plt.legend()

plt.savefig("C:/Users/mohsy/University/KLM/Thesis/My thesis/Plots/Report plots/" 
           + fig_title + '.jpeg')
```
