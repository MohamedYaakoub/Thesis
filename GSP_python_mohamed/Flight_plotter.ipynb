```python
import numpy as np
import seaborn as sns
import pickle
from matplotlib import pyplot as plt
from matplotlib.ticker import MaxNLocator, ScalarFormatter
```

```python
file_name = "Reynolds_CEOD_data_mohamed_2019_feb_1-9_2"
GEnx_OD, GEnx_OD_true, _, alt_time, All_Reynolds = pickle.load(open(f"Reynolds_pickle/{file_name}.p", "rb"))
take_off_All_Reynolds, climb_All_Reynolds, cruise_All_Reynolds = All_Reynolds
Re2, Re25, Re3, Re4, Re49, Re5, Re14, Re19, _, _ = np.array(take_off_All_Reynolds).T
```

```python
fig_title = "Corrected N1 speed"

fig, ax = plt.subplots(2, 1, figsize=(8, 8), dpi=500)
for flight_phase, legend, color in zip(range(3), ['Take off', 'Climb', 'Cruise'], ['#FF1F5B', '#009ADE', '#00CD6C']):

    ax[0].scatter(alt_time[flight_phase][:, 1] / 3600,          GEnx_OD[flight_phase][:, 0], s=2, alpha=0.7,label=legend, c=color)
    ax[1].scatter(alt_time[flight_phase][:, 0] * 0.3048 / 1000, GEnx_OD[flight_phase][:, 0], s=2, alpha=0.7,label=legend, c=color)

    ax[0].set_ylabel(f'Corrected N1 speed to station 12 [%]')
    ax[0].set_xlabel('Time [h]')
    
    ax[1].set_ylabel(f'Corrected N1 speed to station 12 [%]')
    ax[1].set_xlabel('Altitude [km]')
    
    
    
    
ax[0].grid(True, alpha=0.2)
ax[1].grid(True, alpha=0.2)
# ax.set_title(title)

# Use ScalarFormatter for the y-axis
x_formatter = ScalarFormatter(useMathText=True)
x_formatter.set_scientific(True)
# x_formatter.set_powerlimits((2, 2))  # Adjust the power limits for x-axis

y_formatter = ScalarFormatter(useMathText=True)
y_formatter.set_scientific(True)
# y_formatter.set_powerlimits((0, 0))  # Adjust the power limits for y-axis

ax[0].xaxis.set_major_formatter(x_formatter)
ax[0].yaxis.set_major_formatter(y_formatter)
ax[1].xaxis.set_major_formatter(x_formatter)
ax[1].yaxis.set_major_formatter(y_formatter)
fig.tight_layout()
legends = fig.legend(*fig.axes[0].get_legend_handles_labels(), loc='upper right')
for handle in legends.legendHandles:
    handle.set_sizes([10])
    
# plt.savefig("C:/Users/mohsy/University/KLM/Thesis/My thesis/Plots/Report plots/" 
#            + fig_title + '.jpeg')
```
